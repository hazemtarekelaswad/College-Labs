The model has two simple parameters T_a  and T_b .
The two parameters represent unit of time in seconds. The first parameter is the time latency to change the traffic lights from red to green once a car is found waiting in the queue. The second parameter T_b  is the time required for a car to cross the intersection. So once the lights change from red to green, the lights will stay green for an amount of time equals to T_b times the number of cars waiting. The model also assumes that once the lights change from red to green, all cars waiting pass through immediately and at the same time.

```python
from heapq import *
from numpy import random
```

```python
### STATE ##########################################

class State:
    def __init__(self):
        self.green = False
        self.cars = 0
    def is_green(self):
        """
        True if the light is green
        """
        return self.green
    def add_car(self):
        """
        Adds a car in the queue
        """
        self.cars = self.cars + 1
    def purge_cars(self):
        """
        Empty waiting cars
        """
        self.cars = 0
    def waiting_cars(self):
        """
        Returns the number of car waiting
        """
        return self.cars
    def turn_green(self):
        """
        The light turns green
        """
        self.green = True
    def turn_red(self):
        """
        The light turns red
        """
        self.green = False
    def __str__(self):
        """
        Displays the status of the crossroads
        """
        return "Green light =" + str(self.green) + ", cars=" + str(self.cars)
```

```python
### EVENTS ###########################################

Ta = 30 # Time latency to change the traffic lights from red to green once a car is found waiting in the queue
Tb = 10 # Passage time
```

```python
class Event:
    def time(self):
        """
        Returns the time at which the event will be processed
        """
        return self.t
    def __str__(self):
        """
        Displays Event
        """
        return self.name + "(" + str( self.t ) + ")"
    def __lt__(self, other):
        """
        Compares the event with another sorted by processing order priority
        """
        return self.t < other.t
```

```python
class CAR(Event):
    def __init__(self,time):
        self.t = time
        self.name = "CAR"
    def action(self,queue,state):
        if not state.is_green():
            #TODO On this line, change the state and add a car.
            state.add_car()
            if state.waiting_cars() == 1:
                #TODO On this line, insert into the queue a R2G event at the time (self.t + Ta)
                queue.insert( R2G( self.t + Ta ) )
```

```python
class R2G(Event):
    def __init__(self,time):
        self.t = time
        self.name = "R2G"
    def action(self,queue,state):
        queue.insert( G2R( self.t + state.waiting_cars() * Tb ) )
        #TODO On this line, change the state and turn the light to green
        state.purge_cars()
        state.turn_green()
```

```python
class G2R(Event):
    def __init__(self,time):
        self.t = time
        self.name = "G2R"
    def action(self,queue,state):
        #TODO On this line, change the state and turn the light to red
        state.turn_red()
```

```python
### EVENT QUEUE ##############################################

class EventQueue:
    def __init__(self):
        self.q = []
    def notEmpty(self):
        """
        Returns true if the queue is not empty
        """
        return len(self.q) > 0
    def remaining(self):
        """
        Returns the number of events awaiting processing
        """
        return len(self.q)
    def insert(self, event):
        """ 
        Create a new event in the queue
        """
        heappush( self.q, event )
    def next(self):
        """
        Returns and removes from the queue the next event to be processed
        """
        return heappop( self.q )
```

```python
### STATE ##########################################

# states -> open, busy, closed "strings"

class State:
    def __init__(self):
        self.condition = "closed" 
        self.wait_queue = []
        self.total_customers = 0
        self.total_wait = 0
        self.closing_time = 0

    def get_closing_time(self): 
        return self.closing_time
        
    def set_closing_time(self, time): 
        self.closing_time = time

    def is_condition(self):
        return self.condition

    def add_customer(self, customer):
        heappush(self.wait_queue, (customer.t, customer))
    
    def increament_total_customers(self):
        self.total_customers = self.total_customers + 1

    def add_wait_time(self, time):
        self.total_wait += time
        
    def remove_customer(self):
        return heappop(self.wait_queue)
        
    def waiting_customers(self):
        return len(self.wait_queue)
        
    def turn_closed(self):
        self.condition = "closed" 
        
    def turn_busy(self):
        self.condition = "busy" 
        
    def turn_open(self):
        self.condition = "open" 
        
    def __str__(self):
        return "condition =" + str(self.condition) + ", customers=" + str(self.wait_queue)
```

```python
class Event:
    def time(self):
        """
        Returns the time at which the event will be processed
        """
        return self.t
    def __str__(self):
        """
        Displays Event
        """
        return f'{self.name}({self.t})'
    def __lt__(self, other):
        """
        Compares the event with another sorted by processing order priority
        """
        return self.t < other.t
```

```python
class WBUSY(Event):
    def __init__(self, time, duration):
        self.t = time
        self.duration = duration
        self.name = "WBUSY"

    def action(self,queue,state):
        state.add_wait_time(self.t + self.duration) #! Finish Time
        state.turn_busy()
        
        if self.t + self.duration > state.get_closing_time(): 
            queue.insert(WCLOSED(self.t + self.duration))

        elif not state.waiting_customers(): # if it is empty
            queue.insert(WCLOSED(state.get_closing_time()))
            
        queue.insert(INTERNAL_WOPEN(self.t + self.duration))
```

```python
class WCLOSED(Event):
    def __init__(self, time):
        self.t = time
        self.name = "WCLOSED"

    def action(self,queue,state):
        state.turn_closed()
```

```python
class INTERNAL_WOPEN(Event):
    def __init__(self,time):
        self.t = time
        self.name = "INTERNAL_WOPEN"

    def action(self,queue,state):
        if state.is_condition() == "closed": return
        state.turn_open()
        if state.waiting_customers():
            _, customer = state.remove_customer()
            queue.insert(WBUSY(self.t, customer.duration))
```

```python
class WOPEN(Event):
    def __init__(self,time, duration):
        self.t = time
        self.duration = duration 
        self.name = "WOPEN"

    def action(self,queue,state):
        state.turn_open()
        if state.waiting_customers():
            _, customer = state.remove_customer()
            state.set_closing_time(self.t + self.duration)
            queue.insert(WBUSY(self.t, customer.duration))
```

```python
class CUSTOMER(Event):
    def __init__(self, time, duration):
        self.t = time
        self.duration = duration 
        self.name = "CUSTOMER"

    def action(self,queue,state):
        state.increament_total_customers()
        state.add_wait_time(-self.t) #! Arrival Time

        if state.is_condition() == "open":
            state.turn_busy()
            queue.insert(WBUSY(self.t, self.duration))

        else:
            state.add_customer(self)
```

```python
### MAIN #####################################################

Q = EventQueue()

Q.insert( CUSTOMER(0,2) ) #a customer arrives at t=0 , and requires 2 t to finish his processing.
Q.insert( CUSTOMER(1,1) ) 
Q.insert( CUSTOMER(1.1,3) ) 
Q.insert( WOPEN(2,5) )  # the window opens at 2 and closes after 5 t i.e closes at 7 or if it processes a customer then after the customer finishes.
Q.insert( CUSTOMER(2,1) ) 
Q.insert( CUSTOMER(3,1) ) 
Q.insert( WOPEN(10,5) ) 
    
S = State()

# Processing events until the queue is Q is empty
while Q.notEmpty():
    e = Q.next()
    print( e )
    e.action(Q,S)

# print()
# print(f'Total wait time: {S.total_wait}\nTotal number of cutomers: {S.total_customers}')
print(S.total_wait / S.total_customers)
```
