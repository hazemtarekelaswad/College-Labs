```python
import scipy
import random
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def integrate(n, a, b, fx):
    samples = [random.uniform(a, b) for _ in range(n)]
    return (b - a) * sum([fx(sample) for sample in samples]) / n
```

```python
fx = lambda x: ((15*x**3 + 21*x**2 + 41*x + 3)**0.25)*math.exp(-0.5*x)
n = list(range(100, 10001, 100))
a = 1
b = 4
actual_integral, _ = scipy.integrate.quad(fx, a, b)

print(f'Scipy Integration = {actual_integral}')
print(f'MonteCarlo Integration at n of {n[-1]} = {integrate(n[-1], a, b, fx)}')

plt.plot(n, [integrate(size, a, b, fx) for size in n], label='MonteCarlo intergrations')
plt.plot(n, [actual_integral]*len(n), label='Scipy integration')
plt.xlabel('n (sample sizes)')
plt.ylabel('Function integration')
plt.legend()
plt.show()
```

*We can see that the largest sample size (`n`) - which is `10000` in this example - has the closest integration value to the actual result of that integration

```python
n = 1000
a = 1
b = 4
reps = 500
integrations = [integrate(n, a, b, fx) for _ in range(reps)]
plt.hist(integrations, bins=100)
plt.xlabel('integrations')
plt.ylabel('Frequency')
plt.show()

print(f'standard deviation = {np.std(integrations)}')
```

**Notice** *that the for larger n, the less standard deviation will be smaller*
